```python
import numpy as np
import pandas as pd
import requests
```

```python
api_url = f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=USD&order=market_cap_desc&per_page=200&page=1&sparkline=false&price_change_percentage='24h%2C7d%2C14d%2C30d%2C200d%2C1y%2C'"
data = requests.get(api_url).json()
print(data)
```

```python
my_columns = ['Name', 'Ticker','Current Price', 'Market Capitalization']
atl_dataframe = pd.DataFrame(columns = my_columns)
atl_dataframe
```

```python
for i in data:
    name = i['name']
    ticker = i['symbol']
    price = i['current_price']
    market_cap = i['market_cap']
 
    final_dataframe = atl_dataframe.append(
    pd.Series(
        [
            name,
            ticker,
            price,
            market_cap,

            
        ],
            index = my_columns,
    ),
        ignore_index = True
)
atl_dataframe
```
